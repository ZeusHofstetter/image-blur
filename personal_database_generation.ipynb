```python
%matplotlib inline
import matplotlib
import ipywidgets as widgets
from PIL import Image
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.figsize'] = [15,15]
```

```python
from os import listdir

fdir='Data/Personal_ImageBlurDataset/unsorted/'
files= listdir(fdir)
images=[]
for fn in files:
    img_path = fdir+fn
    x=im = Image.open(img_path)
    images.append(x)
```

```python
from IPython.display import display
from IPython.display import clear_output

radio=widgets.RadioButtons(
    options=['default','private', 'undistorted', 'blurred', 'obscured', 'other'],
    value='default',
    description='Image quality:',
    disabled=False
)

caption = widgets.Label(value='0')

plt.imshow(images[0])

def handle_radio_change(change):
    if radio.value != 'default':
        k=int(caption.value)
        if k<len(images):
            w=radio.value
            images[k].save( 'Data/Personal_ImageBlurDataset/'+w+'/'+files[k], 'JPEG' )

            #iterate to next image
            k+=1
            caption.value=str(k)   
            clear_output()
            radio.value='default'
            plt.imshow(images[k])
        else:
            print 'Reached last image'
            return None


radio.observe(handle_radio_change, names='value')


display(radio)
        
        
    
    
    
    
```
